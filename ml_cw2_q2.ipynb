## 2.3 Multilayer perceptron

```python
std_scale = preprocessing.StandardScaler().fit(X_train)
X_train_scaled = std_scale.transform(X_train)

X_test_scaled = std_scale.transform(X_test)

y_train_mlp = (y_train.values - np.mean(y_train)) / np.std(y_train)
y_test_mlp = (y_test.values - np.mean(y_train)) / np.std(y_train)
```

```python
alpha = 1e-2
lr = 1e-4

mlp = tf.keras.Sequential()
mlp.add(layers.Dense(100, activation='relu',
       kernel_regularizer=tf.keras.regularizers.l2(alpha),
       bias_regularizer=tf.keras.regularizers.l2(alpha)))
mlp.add(layers.Dense(1,
       kernel_regularizer=tf.keras.regularizers.l2(alpha),
       bias_regularizer=tf.keras.regularizers.l2(alpha)))

mlp.compile(optimizer=tf.keras.optimizers.Adam(lr),
              loss='mse',
              metrics=[])

# The patience parameter is the amount of epochs to check for improvement
early_stop = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=100)

history = mlp.fit(X_train_scaled, y_train_mlp, epochs=1000, batch_size=64,
       validation_data=(X_test_scaled, y_test_mlp), verbose=0,
                 callbacks=[early_stop, tfdocs.modeling.EpochDots()])
```

```python
plt.figure(figsize=(10,8))
hist = pd.DataFrame(history.history)
hist['epoch'] = history.epoch
plotter = tfdocs.plots.HistoryPlotter(smoothing_std=0)
plotter.plot({'Basic': history}, metric = "loss")
plt.ylabel('MSE [MPG]')
```

```python
y_pred_train_mlp = mlp.predict(X_train_scaled) * np.std(y_train) + np.mean(y_train)
y_pred_test_mlp = mlp.predict(X_test_scaled) * np.std(y_train) + np.mean(y_train)

y_pred_train_mlp_int = np.round(y_pred_train_mlp).astype(int)
y_pred_test_mlp_int = np.round(y_pred_test_mlp.astype(int))
```

```python
print('accuracy TRAIN: ', accuracy_score(y_train, y_pred_train_mlp_int))
print('accuracy TEST: ', accuracy_score(y_test, y_pred_test_mlp_int))
print('\n')
print('mse train: ', mean_squared_error(y_train, y_pred_train_mlp))
print('mse test: ', mean_squared_error(y_test, y_pred_test_mlp))
print('\n')
print("r2 train : ", r2_score(y_train, y_pred_train_mlp))
print("r2 test : ", r2_score(y_test, y_pred_test_mlp))
```

```python
f, (ax1, ax2) = plt.subplots(ncols=2, figsize=(15,6))
ax1.scatter(y_train, y_pred_train_mlp_int)
ax1.plot([1922,2011],[1922, 2011], color='black', linestyle='--')
ax1.set_xlabel('TRAIN true year')
ax1.set_ylabel('TRAIN predicted year')
ax2.scatter(y_test, y_pred_test_mlp_int)
ax2.plot([1922,2011],[1922, 2011], color='black', linestyle='--')
ax2.set_xlabel('TEST true year')
ax2.set_ylabel('TEST predicted year')
plt.show()
```

### Cross Validation

```python
class NN():
    def __init__(self, n_neurons, lr, alpha, epochs=1000, batch_size=64):
        self.alpha = alpha
        self.lr = lr
        self.epochs=epochs
        self.batch_size=batch_size

        self.mlp = tf.keras.Sequential()
        for n in n_neurons:
            self.mlp.add(layers.Dense(n, activation='relu',
               kernel_regularizer=tf.keras.regularizers.l2(self.alpha),
               bias_regularizer=tf.keras.regularizers.l2(self.alpha)))
            
        self.mlp.add(layers.Dense(1,
           kernel_regularizer=tf.keras.regularizers.l2(self.alpha),
           bias_regularizer=tf.keras.regularizers.l2(self.alpha)))

        self.mlp.compile(optimizer=tf.keras.optimizers.Adam(self.lr),
                  loss='mse',
                  metrics=[])
        
    def fit(self, X,y, X_test=None, y_test=None):
        early_stop = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=100)
        if X_test is None:
            self.history = self.mlp.fit(X, y, epochs=self.epochs, batch_size=self.batch_size,
                                        callbacks=[tfdocs.modeling.EpochDots()],
                                        verbose=0)     
        else:
            self.history = self.mlp.fit(X, y, epochs=self.epochs, batch_size=self.batch_size,
                            validation_data = (X_test, y_test),
                            callbacks=[early_stop,tfdocs.modeling.EpochDots()],
                            verbose=0)
        
    def predict(self, X):
        return self.mlp.predict(X)
```

```python
params_grid = {'n_neurons': [[100], [50, 30, 10], [10 for k in range(8)]],
               'lr': np.logspace(-5, 0, 5),
               'alpha': np.logspace(-4, 0, 5)}
```

```python
best_params = {'n_neurons': [100], 'lr': 0.000177828, 'alpha': 0.01}
```

```python
best_nnet = NN(**best_params)
best_nnet.fit(X_train_scaled, y_train_mlp, X_test_scaled, y_test_mlp)
```

```python
y_pred_train_mlp = best_nnet.predict(X_train_scaled) * np.std(y_train) + np.mean(y_train)
y_pred_test_mlp = best_nnet.predict(X_test_scaled) * np.std(y_train) + np.mean(y_train)

y_pred_train_mlp_int = np.round(y_pred_train_mlp).astype(int)
y_pred_test_mlp_int = np.round(y_pred_test_mlp.astype(int))

print('accuracy TRAIN: ', accuracy_score(y_train, y_pred_train_mlp_int))
print('accuracy TEST: ', accuracy_score(y_test, y_pred_test_mlp_int))
print('\n')
print('mse train: ', mean_squared_error(y_train, y_pred_train_mlp))
print('mse test: ', mean_squared_error(y_test, y_pred_test_mlp))
print('\n')
print("r2 train : ", r2_score(y_train, y_pred_train_mlp))
print("r2 test : ", r2_score(y_test, y_pred_test_mlp))
```
